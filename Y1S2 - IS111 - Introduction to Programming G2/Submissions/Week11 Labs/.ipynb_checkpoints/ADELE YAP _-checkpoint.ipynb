## Q1: Substitution Cipher

"A __substitution cipher__ is a method of encoding by which units of plaintext are replaced with ciphertext, according to a fixed system." See https://en.wikipedia.org/wiki/Substitution_cipher.

In this question, we consider a simple substitution cipher that replaces each letter in the original message with another letter.

### Part (a) [ &bigstar; ]

You are given a dictionary that stores the mappings from English lowercase letters to their replacements. Note that no two letters will be mapped to the same replacement.

Define a function called `encrypt()` that takes in two parameters:

- `my_dict` (type: `dict`): a dictionary as described above. 
- `msg` (type: `str`): a message consisting of only lowercase letters and spaces.

The function returns the encrypted message using `my_dict` as the substitution cipher. Note that a space remains untouched in the encrypted message.

````python
# Write your function definition here.

def encrypt(cipher_dict, message):
    # cipher_dict = {} --> not needed as it is already a parameter in the qns
    ## start off with an empty str 
    encrypted_message = ''     

    ## need to iterate thru all the characters in the message 
    for ch in message:

        ## checks if the character isnt an empty space
        # if != empty space it will retreive the corresponding cipher text from the dict using cipher_dict[ch] 
        # store in encrypted_ch 
        if ch != ' ':   ## needs to be ' ' and not '' cuz the key '' (empty str isnt found in the cipher dict)
            encrypted_ch = cipher_dict[ch]
        
        ## if the ch is an empty space then it will be stored as an empty space in the encrypted ch 
        else:
            encrypted_ch = ch

        ## append encrypted ch to encrypted message str --> loop will move on to next ch 
        encrypted_message += encrypted_ch
        ## needs to have the same level of indentation as "if ch != ' ': and else: " to be part of the same block and executed inside the for loop --> build full encrypted msg ch by ch 
    
    ## once entire msg crypted --> return full encrypted message str 
    return encrypted_message
````

````python
def encrypt(cipher_dict, message):
    encrypted_message =''
    for ch in message:
        if ch != ' ':
            encrypted_ch = cipher_dict[ch]
        else: 
            encrypted_ch = ch 
        encrypted_message += encrypted_ch
    return encrypted_message
````

### Part (b) [ &bigstar;&bigstar; ]

Now define a function called `decrypt()`. This function also takes in two parameters:

- `my_dict` (type: `dict`): a dictionary as described above. 
- `encrypted_msg` (type: `str`): an __encrypted__ message.

The function returns the original message, i.e., the message that has been encrypted into the string `encrypted_msg`.

````python
# Write your function definition here.
def decrypt(my_dict, encrypted_msg):

    new_dict = {}   ## to store reversed key-value pairs from my_dict

    ## iterate thru each key-value pair in 'mydict' 
    for key in my_dict:
        ## value becomes key and key becomes value since placements are swapped 
        value = my_dict[key]
        new_dict[value] = key 

    ## create an empty str to store the reversed msg
    decrypted_msg = ''
    ## iterate thru the ch in encrypted msg 
    for ch in encrypted_msg:
        ## if ch is a space, then it will remain a space in the decrypted msg
        if ch == ' ': 
            decrypted_msg += ' '
        ## if ch is not a space and is a ch in the encrypted msg 
        else:  
            # corressponding decrypted ch is obtained from newdict dictionary by looking up value of 'ch' 
            # store in 'newch' variable 
            new_ch = new_dict[ch]
            ## append decrypted new ch element to decrypted msg str 
            decrypted_msg += new_ch

    ## once all ch have been sifted thru, return the full decrypted msg str 
    return decrypted_msg
````

````python
def decrypt(my_dict, encrypted_msg):
    new_dict = {}
    for key in my_dict:
        value = my_dict[key]
        new_dict[value] = key
    
    decrypted_msg = ''
    for ch in encrypted_msg:
        if ch == ' ':
            decrypted_msg += ' '
        else: 
            new_ch = new_dict[ch]
            decrypted_msg += new_ch
    return decrypted_msg
````

## Q2: Prime Numbers [ &bigstar;&bigstar; ] 


Define a function called ```create_prime_dict()``` that takes in a list of integers greater than 1 (without any duplicates) as its parameter. The function creates a dictionary that maps each integer in the list to a boolean value that indicates whether or not this integer is a prime number.

Recall that a prime number is a number that is divisible only by 1 and itself.

For example, ```create_prime_dict([2, 3, 5, 10, 20, 23])``` returns a dictionary that looks like the following: ```{2:True, 3:True, 5:True, 10:False, 20:False, 23:True}```.

````python
# Put your code here:

##check if int is a prime num or not first 
## if is prime == return True, else return False
def is_prime(num):
    #iterate thru all nums to get prime nums
    for i in range(2, num):
        # if num is divisible by i, it cant be prime = false
        if num % i == 0: 
            return False 
    return True     ## else = True 

def create_prime_dict (int_list):
    ## int_list = [] --> no need create empty list cuz it overrides input parameter
    my_dict = {}

    ##iterate thru all the nums in the int list 
    for num in int_list:
        ## check if num in int list is a prime num
        if is_prime(num): 
            ## if it is a prime num, then the value in the dict == prime --> return true
            my_dict[num] = True
        else: 
            ## if not prime, then false 
            my_dict[num] = False 
    return my_dict
````

````python
def is_prime(num):
    for i in range(2,num):
        if num % i == 0:
            return False 
    return True 

def create_prime_dict(int_list):
    my_dict = {}
    for num in int_list:
        if is_prime(num):
            my_dict[num] = True
        else:
            my_dict[num] = False 
    return my_dict
````

### Part (a) [ &bigstar;&bigstar; ]

Write a program that allows a user to continuously check the details of a room until the user chooses to quit the program.

A sample run of the program may look like the following:

```
Do you want to search for a facility? [Y|N] :Y
Enter the school [LKCSB|SOE|SOL|SOA|SIS] :SOE
```

````python
# Put your code here:
facilities_dict = {}

with open('facilities.txt', 'r') as input_file:
    for line in input_file:
        column = line.rstrip('\n').split('\t')
        school = column[0]
        room_number = column[1]
        capacity = int(column[2])
        has_projector = column[3]

        key = (school, room_number)
        value = (capacity, has_projector)
        facilities_dict[key] = value

is_searching = input ("Do you want to search for a facility? [Y|N] :")
while is_searching == 'Y':
    school= input("Enter the school [LKCSB|SOE|SOL|SOA|SIS] :")
    room_number= input("Enter the room number :")

    key = (school, room_number)
    value = facilities_dict[key]
    facilities_dict[key] = value 
    capacity = value[0]
    has_projector = value[1]

    if has_projector == 'yes': 
        has_projector = 'has a projector'
    else:
        has_projector = 'does not have a projector'
    print ('Do you want to search for a facility? [Y|N] :' + is_searching)
    print("Enter the school [LKCSB|SOE|SOL|SOA|SIS] :" + school)
    print ("Enter the room number :" + room_number)
    print (school, room_number, 'has a capacity of ', capacity, 'and ', has_projector, '.' , '\n' )

    is_searching = input ("Do you want to search for a facility? [Y|N] :")

print ('Do you want to search for a facility? [Y|N] :' + is_searching)
print ("Thanks for using our system!")
````

### Part (a) [ &bigstar;&bigstar;&bigstar; ]

Define a function called ```count_words_in_file()```. The function takes in as its parameter the name of a file, where each line of the file is the content of one article.

The function returns a dictionary that stores the mappings from all the unique words found in the file to their total counts in the file.

For example, if the word "the" has appeared 10 times the file, then the mapping ```'the':10``` should be stored inside the dictionary.

You need to turn everything into lowercases before counting. For example, "The" and "the" are both counted as ```'the'``` in the dictionary.

Simply use ```line.split()``` to break down a line into a list of words. You do not need to handle punctuation marks for this question.

````python
# Put your code here:
def count_words_in_file (file):
    #filename = line is the content of one article 
    # func returns dict that stores mappings from all unique words found in file to their total counts in the file
    my_dict = {}   #key = word, value = num of times it appears 
  
    with open (file, 'r') as my_file:       ##rem to use my_file! 
        for line in my_file:
            unique_words = line.rstrip('\n').split()     # break down line into list of words 
            for unique_word in unique_words:    ##break down the list of words into a singular word 
                lower_unique_word = unique_word.lower()     #make each word lowercased 
                if lower_unique_word in my_dict:        # if word is found in dict then need to add 1 to its value 
                    my_dict[lower_unique_word] += 1
                else: 
                    my_dict[lower_unique_word] = 1      ##if word not in dict, add new keyvalue pair to dict --> key = lowercase word, value initialised to 1 
                
                ## used to count the occurrences of each unique word in the file. 
                # If the word is already in the dictionary, it means that it has already been encountered before, so the count is incremented. 
                # If it is not in the dictionary, it means that it is a new word that has not been encountered before, so a new key-value pair is added to the dictionary with a count of 1.

    return my_dict
````

````python
def count_words_in_file (file):
    my_dict = {}  

    with open (file, 'r') as my_file:
        for line in my_file:
            unique_words = line.rstrip('\n').split()    
            for unique_word in unique_words:
                lower_unique_word = unique_word.lower()
                if lower_unique_word in my_dict:
                    my_dict[lower_unique_word] += 1
                else: 
                    my_dict[lower_unique_word] = 1
    return my_dict
````
