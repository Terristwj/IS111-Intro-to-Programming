### Q3 [de morgan's]
Which of the following options always evaluates to False?

```
A. not ( not (a and b) or (not a or b) )
B. not ( not (a or b) or (not a or b) )
C. Both choices always evaluate to `False`
D. Neither choice always evaluates to `False`
```

#### Solution:
```
B:
False ?= not ( not (a or b) or (not a or b) )
not (False) ?= not( not ( not (a or b) or (not a or b) ) )
True ?= not (a or b) or (not a or b)
True ?= not a and not b or not a or b
```

| `a` | `b` | `not a and not b or not a or b` | Result |
|-----|-----|---------------------------------|--------|
|  0  |  0  | `not F and not F or not F or F` | `True` |
|  0  |  1  | `not F and not T or not F or T` | `True` |
|  1  |  0  | `not T and not F or not T or F` | `False`|
|  1  |  1  | `not F and not F or not F or F` | `True` |

```
A:
False ?= not ( not (a and b) or (not a or b) )
not (False) ?= not( not ( not (a and b) or (not a or b) ) )
True ?= not (a and b) or (not a or b)
True ?= not a or not b or not a or b
True ?= not a or not b or b
```

From OR Truth Table, `b and not b` must always be `True`
Hence, this is always `True`

```python
def modify_string(input_str):
	last_index = 0
	output = ''
    # iterate over the position indices of input_str
	for current_index in range(len(input_str)):
		
        # only do something if current position is a space (' '); otherwise continue
        if input_str[current_index] == ' ':
            
            # word is an input_str sliced from position index "last_index" to "current_index"
			word = input_str[last_index:current_index]
            
            # if word has more than 5 characters
            if len(word) > 5: 
                # do the next step (len(word) - 3)//2 times; i.e., 6 => 2x, 7 => 2x, 8 => 3x, etc.
				for i in range(3, len(word), 2):
                    # concatenate '*' to output; number of '*" depends on len(word)
					output = output + '*'                   
			else:
                # concatenate word to output
				output = output + word

            # concatenate space to output
			output = output + ' '
            # move "last_index" to the position index after the space
			last_index = current_index + 1
	return output
```
