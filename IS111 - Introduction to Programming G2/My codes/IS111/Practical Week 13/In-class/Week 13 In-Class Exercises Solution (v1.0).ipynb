### Q1: Count Strings [ &bigstar;&bigstar; ] 

```python
# SOLUTION

str_input = input('Enter a string > ')
str_dict = {}

while len(str_input) != 0:
    first_char = str_input[0].lower()
    if first_char not in str_dict:
        str_dict[first_char] = 1
    else:
        str_dict[first_char] += 1
    
    str_input = input('Enter a string > ')

print("\nYou've entered")
for key in str_dict:
    count = str_dict[key]
    
    # handle "strings" vs "string"
    if count > 1:
        word_str = ' strings'
    else:
        word_str = ' string'
        
    print('\t' + str(count) + word_str + " starting with '", end='')
     
    # If the key is a letter, then print an extra uppercase letter.
    if key.lower() in 'abcdefghijklmnopqrstuvwxyz':
    # Or you can use the following code:
    # if key.isalpha():
        print(key.upper() + "' or '", end='')
        
    print(key + "'")
```

### Q2: Count Emails  [ &bigstar;&bigstar; ] 

```python
# Solution 1

def convert_timestamp(timestamp):
    # This function takes in a parameter that is a string representing a timestamp.
    # The timestamp has one of the following formats:
    #      - 'hh:mm:ss' where hh is between 00 and 23
    #      - 'hh:mm:ssAM' where hh is between 01 and 12
    #      - 'hh:mm:ssPM' where hh is between 01 and 12
    # The function returns a tuple consisting of three integers representing the hours, 
    # minutes and seconds of this timestamp.
    
    is_am = False
    is_pm = False
    if len(timestamp) == 10:
        # Check if the timestamp is an 'AM' or 'PM' timestamp
        if timestamp[8] == 'A':
            is_am = True
        else:
            is_pm = True
        # Remove 'AM' or 'PM' at the end.
        timestamp = timestamp[0:8]
    
    # Split the timestamp into hh, mm and ss.
    (hour, minute, second) = timestamp.split(':')
    # Convert everything into integers.
    hour = int(hour)
    minute = int(minute)
    second = int(second)
    
    # Handle AM and PM timestamps.
    if is_am and hour == 12:
        hour = 0
    elif is_pm and hour < 12:
        hour += 12
    
    return (hour, minute, second)
    
def compare_time(time1, time2):
    # This function takes in two parameters. Each parameter is a tuple consisting of three integers 
    # representing hours, minutes and seconds.
    # The function returns -1 if time1 is before time2, 0 if time1 is the same as time2, and 1 if 
    # time1 is after time2.
    
    hour1 = time1[0]
    hour2 = time2[0]
    minute1 = time1[1]
    minute2 = time2[1]
    second1 = time1[2]
    second2 = time2[2]
    
    
    if hour1 < hour2:
        return -1
    elif hour1 > hour2:
        return 1
    else:
        if minute1 < minute2:
            return -1
        elif minute1 > minute2:
            return 1
        else:
            if second1 < second2:
                return -1
            elif second1 > second2:
                return 1
            else:
                return 0

def count_emails(file_name, start_time, end_time, sender):
    count = 0
    
    # Convert start_time and end_time to tuples with three integers, where the last integer
    # represents seconds. This is to make their representations consistent with
    # how we will represent an email's timestamp later.
    start_time = (start_time[0], start_time[1], 0)
    end_time = (end_time[0], end_time[1], 0)
    
    with open(file_name, 'r') as input_file:
        for line in input_file:
            line = line.rstrip('\n')
            cols = line.split('\t')
            
            my_timestamp = cols[0]
            my_sender = cols[1]
            
            if my_sender == sender:
                
                # Convert the timestamp from a string to a tuple with three integers.
                my_time_tuple = convert_timestamp(my_timestamp)
                
                if compare_time(my_time_tuple, start_time) >= 0 and compare_time(my_time_tuple, end_time) <= 0:
                    count += 1
    
    return count
```

```python
# Solution 2

def get_time_in_seconds(timestamp):
    # This function takes in a parameter that is a string representing a timestamp.
    # The timestamp has one of the following formats:
    #      - 'hh:mm:ss' where hh is between 00 and 23
    #      - 'hh:mm:ssAM' where hh is between 01 and 12
    #      - 'hh:mm:ssPM' where hh is between 01 and 12
    # The function returns an integer representing the number of seconds that have passed
    # since 00:00:00 corresponding to this timestamp.
    
    is_am = False
    is_pm = False
    if len(timestamp) == 10:
        # Check if the timestamp is an 'AM' or 'PM' timestamp
        if timestamp[8] == 'A':
            is_am = True
        else:
            is_pm = True
        # Remove 'AM' or 'PM' at the end.
        timestamp = timestamp[0:8]
    
    # Split the timestamp into hh, mm and ss.
    (hour, minute, second) = timestamp.split(':')
    # Convert everything into integers.
    hour = int(hour)
    minute = int(minute)
    second = int(second)
    
    # Handle AM and PM timestamps.
    if is_am and hour == 12:
        hour = 0
    elif is_pm and hour < 12:
        hour += 12
    
    return hour * 60 * 60 + minute * 60 + second
    

def count_emails(file_name, start_time, end_time, sender):
    count = 0
    
    start_time_in_sec = start_time[0] * 60 * 60 + start_time[1] * 60
    end_time_in_sec = end_time[0] * 60 * 60 + end_time[1] * 60
    
    with open(file_name, 'r') as input_file:
        for line in input_file:
            line = line.rstrip('\n')
            cols = line.split('\t')
            
            my_timestamp = cols[0]
            my_sender = cols[1]
            
            if my_sender == sender:
                
                my_timestamp_in_sec = get_time_in_seconds(my_timestamp)
                
                if (my_timestamp_in_sec >= start_time_in_sec and my_timestamp_in_sec <= end_time_in_sec):
                    count += 1
    
    return count
```
