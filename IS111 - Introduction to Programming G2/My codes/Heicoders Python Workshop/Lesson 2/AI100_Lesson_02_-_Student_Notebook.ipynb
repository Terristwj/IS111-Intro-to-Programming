1a. Why are Collections Necessary?
---

There are times when we need to store multiple values in a single entity because:
- they are all subset of one entity (e.g. namelist), and logically belong together
- we want to group values together for ease of access

<img src='
' width='550' height='350'/>



```python
# Isn't it inconvenient that we have to print each of the sales value one by one even though they are related?
video_sales_2008 = 1000
video_sales_2009 = 1500
video_sales_2010 = 2000

print(video_sales_2008)
print(video_sales_2009)
print(video_sales_2010)
```

<div class="alert alert-block alert-warning">
To solve the above problem, we can make use of a data type call <b>collections</b>
</div>

1b. What are Collections?
---

A <b>collection is a data type</b> that allows you to carry many values together in one single package. There are several types of collection:

- list
- tuple
- dictionary
- set




Regardless of the differences between the different types of collection, the concept and idea behind them are similar. The following shows examples of a commonly-used type of collection (<b>list</b>):

```python
# As you can see here, we bundle the sales in years 2008, 2009, & 2010 into a single entity
[1000, 1500, 2000]
```

Any <b>values between the 2 square brackets</b> are values of a list. <b>List can be stored in a variable</b> for ease of access.

```python
annual_sales = [1000, 1500, 2000]
print(annual_sales)
```

---
At this point we will make an outright claim that <b> collections are python's representation of databases / tabular data (any form of data in table form such as excel of spreadsheet).</b> Thus, as an aspiring data scientist it is highly important to have mastery of collections.

At the end of this session, we will substantiate our claims once you gain a better appreciation of the mechanism behind collections.

2.Lists
===
***

2a. What is a List?
---

A list is:
- a compound data type that contains <b>an indexed collection of data values.</b> 
- Using a real world analogy, you can liken this system to a <b>list of student names with a register number tagged>/b> to each name.

```python
# Examples of list data type
annual_sales = [1000, 1500, 2000]
year = ['2008', '2009', '2010']
```

2b. How to Create a List?
---

The way to create a new list is:
- Using the <code>[ ]</code> operator
- If there are no values between the square brackets, we term it an <b>empty list</b>
- <b>Any values (as long as it is a python data type e.g. int, float, string)</b> you put between the square brackets would constitute as part of the list.

```python
# This is how you create a list
row_data = [1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49]
print(row_data)
```

<div class="alert alert-block alert-warning">
Notice how you can have <b>multiple data types</b> within the same list? And observe that the sample list is above actually representative of one row of data in the video sales table</div>

2c. How to Access the Data in the List?
---
Similar to strings, <b>lists have an indexed system</b> to keep track of every value stored within. There are two ways which you can extract data from a list:

i) Single value index <br> 
ii) List slicing

<b>2c.i) Single value index</b>

Similar to strings, you can simply pass in the index number which the data within the list is tagged to, in order to extract the data

<img src='
' width='600' height='300'/>


```python
## Extract a particular data point from a list
row_data = [1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49]
print(row_data[2])
```

You can also <b>store the extracted value into a variable</b>

```python
## Store an extracted data point into a variable
row_data = [1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49]

platform = row_data[2]
print(platform)
```

<b>2c.ii) List slicing</b>

Lastly, similar to strings, you can also perform slicing with a list. When you slice a list, you will be <b>returned with a list populated with the values specified by the index range </b>.

<img src='
' width='600' height='300'/>

```python
# To extract a subset of a list
row_data = [1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49]
print(row_data[0:3])
```

Just like for strings, there are many other ways to slice a list.

```python
print(row_data[:3])
```

```python
print(row_data[::-1])
```

2d. How to Update the Data in the List?
---
To assign a new value to a list, you need to make use of the assignment operator <code>=</code> and the index of the list




```python
# To update the data in a list
row_data = [1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49]
row_data[6] = 100
print(row_data)
```

<div class="alert alert-block alert-warning">
Note: Lists are <b>mutable</b> - meaning that we can change the values of the element within the list

Note that <code>remove()</code>:

```python
sample_data = [10, 20, 30, 40, 50, 30]
print(sample_data)

# Remove values from sample_data
sample_data.remove(30)
print(sample_data) # Notice that only one instance of '30' is removed from the list
```

3.Tuples [Self-Reading]
===
***

3a. What is a Tuple?
---

A tuple is:
- a compound data type that contains <b>an indexed collection of data values.</b> 
- Tuples are very similar to list <b>except that they are immutable</b>.



```python
# Examples of tuple data type
game_sales = (1, 'Wii Sports', 'Wii', 2006)
missing_values = ('Sports', 'Nintendo', 41.49)
```

<div class="alert alert-block alert-warning">
In programming when we say an object is immutable, this means that <b>its state cannot be modified after it is created</b>.</div>

3b. How to Create a Tuple?
---

The way to create a new tuple is:
- using the <code>( )</code> operator. 
- If there are no values between the square brackets, we term it an <b>empty tuple</b>. 
- <b>Any values (as long as it is a python data type e.g. int, float, string)</b> you put between the round brackets would constitute as part of the tuple. 
- <b>Visually, there is not much difference between a tuple and list.</b>


```python
# This is how you create a tuple
row_data = (1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49)
print(row_data)
```

<div class="alert alert-block alert-warning">
Note: Notice how you can have multiple data types within the same tuple? And observe that this sample tuple is actually representative of one row of data in the video sales table.
</div>

3c. How to Access the Data in the Tuple?
---
Similar to lists, <b>tuples have an indexing system</b> to keep track of every value stored within. There are two ways which you can extract data from a tuple:

i) Single value index <br> 
ii) Tuple slicing


<b>3c.i) Single value index</b>

Similar to strings, you can simply pass in an index number of the data within the tuple to extract the data

```python
# To extract a particular data from the tuple that was created..
row_data = (1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49)
print(row_data[2]) #Note that the way to access data from a tuple is exactly the same as a list
```

You can also <b>store the extracted value into a variable</b>

```python
# To extract a particular data from the list and store in a variable..
row_data = (1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49)
platform = row_data[2]
print(platform)
```

<b>3c.ii) Tuple slicing</b>

Lastly, similar to string & list, you can also perform slicing with a tuple. When you slice a tuple, you will be <b>returned with a tuple populated with the values specified by the index range</b>.

```python
# To extract a subset of a tuple
row_data = (1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49)
print(row_data[0:3])
```

Just like string & list, there are multiple ways to slice a tuple

```python
print(row_data[:3])
```

```python
print(row_data[::-1])
```

3d. Tuples are immutable!
---
The key difference between a tuple and list is that you <b>cannot change the contents of a tuple once it is created!</b>

```python
## Note: this cell throws a TypeError since tuples are immutable.

# Let's try to update the data in a tuple
row_data = (1, 'Wii Sports', 'Wii', 2006, 'Sports', 'Nintendo', 41.49)
row_data[6] = 100
print(row_data)
```

<div class="alert alert-block alert-warning">
Note: This property is what makes <b>accessing data from tuples faster than list</b> (common question during data science interviews). But you should only implement your database with tuples if you are sure you do not need to change the contents</div>

This also means that the following list functions <b>does not work for Tuples</b>. They will throw you an error.
- append
- extend / concatenation
- remove
- pop
- sort

4a. What is a Dictionary?
---

A dictionary is:
- a compound data type that contains <b>a collection of key-value pairs.</b>
- The concept of python dictionary is similar to that of an Oxford dictionary where every word is unique and has a meaning behind it. 
- Similarly, in python dictionary the key is unique and is tagged to a value. We term this a <b>key-value pair</b>.





```python
# Example of dictionary data type
annual_sales = {'2008': 1000, '2009': 1500, '2010': 2000}
```

<div class="alert alert-block alert-warning">

6b. Comparison Operators
---


Comparison operators <b>evaluate if an expression is true or false</b>.


|Operator | Example | Output |
| ---------|---------|---------|
| `>` | <center>x > y</center>| Evaluates to <b>True</b> if x is greater than y |
| `>=` | <center>x >= y</center>| Evaluates to <b>True</b> if x is greater than or equal to y |
| `<` | <center> x < y </center>| Evaluates to <b>True</b> if x is lesser than y |
| `<=` | <center>x <= y</center>| Evaluates to <b>True</b> if x is less than or equal to y |
| `==` | <center>x == y</center>| Evaluates to <b>True</b> if x is equal to y |
| `!=` | <center>x != y</center>| Evaluates to <b>True</b> if x is not equal to y |

```python
# Example
ten = 10
twenty = 20
thirty = 30

print(ten > twenty)
print(ten < twenty) # Notice that the comparison either evaluates to True or False
```

6i. elif Statement
---

Again, you might realise that there is a <b>limitation with just knowing the <code>if</code> and else <code>else</code> statement</b>. This construct is still too limited in the sense that it only allows you a binary pathway.


Here is the <b>intuition for why we need more than just a <code>if</code> and else <code>else</code> statement</b>:
- Imagine you were tasked to write the function to determine what level of subsidies to provide to people,
- and this subsidies depends on the income level which can be split into 3 range. 
- In that case, is a binary pathway sufficient to determine the subsidies which a person qualifies for?

Hence there is actually another construct call the <code>elif</code> statement</b> that allows you to evaluate other conditions when the <code>if</code> statement</b> condition is false: 




Before we begin, let's see an example of <b>how we would perform a task without the use of the <code>elif</code> statement</b>. This will allow you to appreciate the use of this construct later on. Imagine we were writing an application where the user will key in their salary, and the code will print out the amount of government grant they should receive. In this example:

- If a citizen earns less than 30000 annually, they should receive 5000 in grant <br>
- If a citizen earns more than 30000 but less than 70000 annually they should receive 2000 in grant
- If a citizen earns more than 70000 annually, the receive no grant

<b>Without the use of the <code>elif</code></b>

<b>Using the <code>if</code>, <code>else</code> and <code>elif</code> statement :</b>

```python
salary = 20000

if salary < 30000:
    print('You should get $5000 in grant')
    
if salary >= 30000 and salary < 70000: 
    print('You should get $2000 in grant')
    
if salary > 70000:
    print('You get no grant')
    
```

```python
salary = 60000

if salary < 30000:
    print('You should get $5000 in grant')
    
elif salary >= 30000 and salary < 70000: 
    print('You should get $2000 in grant')
    
else:
    print('You get no grant')
    
```

Let's compare the codes side by side to better understand why <b>using the if-elif-else construct is more efficient</b>!

<div class="alert alert-block alert-warning">
In this above if-elif-else code block, python will stop evaluating the conditons when it evaluated the second condition to be true. In a way, the algorithm is now much more intelligent and efficient. 
</div>

There is <b>still some minor improvement which we can make to the algorithm</b> to make it more succinct.Let's compare the two codes side by side:

<div class="alert alert-block alert-warning">
    Both versions are the <b>same in outcome and logic</b>, the only difference is that one version is shorter.
</div>

Below is the implementation of the code

```python
salary = 60000

if salary < 30000:
    print('You should get $5000 in grant')
    
elif salary < 70000: 
    print('You should get $2000 in grant')
    
else:
    print('You get no grant')
    
```

Note that you can <b>evaluate more than 3 conditions by having more than one elif</b> in the same code block.

```python
salary = 80000

if salary < 30000:
    print('You should get $5000 in grant')
    
elif salary < 70000:
    print('You should get $2000 in grant')

elif salary < 90000:
    print('You should get $1000 in grant')
    
else:
    print('You get no grant')
```

7a. Need for and Implementation of Nested Decisions
---

With the use of <code>if</code>, <code>else</code> and <code>elif</code> statements:
- we can now check for as many criteria as we want. But sometimes, there are computer programs that require more sophistication, where we wish to check for <b>further conditions only when prerequisite conditions are met</b>.
- For instance, imagine if the government wants to further refine the subsidies program. 
- Now, after you meet the salary requirements, they want to see if you meet the age requirements. 

To check for this, you could make use of <b>nested decisions</b> (which involves nesting your conditions):

<b>Without Nested Decisions</b>

<b>With Nested Decisions</b>

```python
salary = 70000
age = 55

if age > 50 and salary < 30000:
 
    print('You should get $5000 in grant')
    
elif age > 50 and salary <= 70000: 
    
    print('You should get $2000 in grant')
    
else:
    print('You get no grant')
```
