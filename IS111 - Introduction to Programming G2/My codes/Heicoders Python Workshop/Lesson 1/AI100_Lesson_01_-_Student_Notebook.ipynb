3h. Comments
---

Other than ignoring blank space, python also ignores any sequence of characters that <b>appears to the right of</b> <code>#</code> symbol. In the example below, we use <code>#</code> before <code>print(1 + 2)</code>, and we see the output of <code>print(1 + 2)</code> is not displayed anymore — this is because <code>print(1 + 2)</code> is<b> not executed when it's preceded by</b> <code>#</code>.

We can either add comments:<br>
i) To ignore specific lines of code<br>
ii) Add specific description to a line of code<br>
iii) Add a general description at the beginnning of our programme

<b> 3h.i) To ignore specific lines of code</b>

````python
# print(1 + 2)
print(10 + 10)
````

<b>3h.ii) Add specific description to a line of code</b>

````python
print(1 + 2)
print(10 + 10) # This line of code outputs 20
````

4e. Rules for Variable Names
---

While we can arbitrarily choose a name for the variables, there are actually <b>3 rules</b> with regard to variable names:

i) The variable name must follow the <b>syntax rules</b> <br>
ii) You cannot use <b>reserved keywords</b> for your variable name <br>
ii) You should use variables that makes sense (this is more of a guide than an actual rule)

<b>4e.i) The variable name must follow the syntax rules</b>


These are the two syntax rules we need to be aware of when we're naming variables:

- We must use only letters, numbers, or underscores (we can't use apostrophes, hyphens, whitespace characters, etc.).
- Variable names cannot start with a number



Here are examples of valid variable names:

````python
## Valid variable names

date_12345 = 1
RESULT_1 = 2
Result_1 = 3
ouTput = 4
_Output = 5

print(date_12345)
print(RESULT_1)
print(Result_1)
print(ouTput)
print(_Output)
````

Here are examples of invalid variable names:

````python
## Invalid variable names (what's wrong with each of them?)

1_data = 1
new data = 2
john's_salary = 3
old-data = 4
price_in_$ = 5
````

<b>4e.ii) Reserved keywords cannot be used as variable name</b>

Reserved words (also called keywords) are defined with predefined meaning and syntax in the language. These <b>keywords have to be used to develop programming instructions</b>. Reserved words can’t be used as identifiers for other programming elements like name of variable, function etc.

````python
#The following are the list of keywords that you cannot use as variables names
#They have been reserved for python

import keyword
keyword.kwlist
````

<b>4e.iii) You should use variable names that are self-descriptive (this is more of a guide than an actual rule)</b>

- We should name variables in a way that will help us remember what we intend to store in them
- This will also help other programmers who are looking at your code understand what you are doing

This is a negative example of variable naming:

````python
## Negative example of variable naming

a = 20
b = 5
c = a * b
print(c)
````

This is a positive example of variable naming:

````python
## Positive example of variable naming

pay_rate = 20
hour = 5
pay = pay_rate * hour
print(pay)
````

While both examples derive the same output, it is quite evident which one makes for easier understanding.

4f. Updating the Values in Variables
---

You can <b>reassign new data</b> to a variable that already has a stored data. The are two forms where this can be done:

i) Update a variable with another value,<br>
ii) Update a variable by performing arithmetic operations

The following shows the output you should get when you assign a value to a variable. Observe the difference with the subsequent examples.

````python
# This is what you should get whem there is no reassignment
result = 20
print(result)
````

<b>4f.i) Update a variable with another value</b>


````python
# Observe the output when a new value is reassigned to the variable
result = 20
print(result)

result = 40
print(result)
````

<b>4f.ii) Update a variable by performing arithmetic operations</b>

````python
# Observe the output when a new value is reassigned to the variable
result = 20

print(result + 30)
print(result)

result = result + 30    # What happened here?
print(result)
````

Notice in the code above that:
- The variable <code>result</code> initially stores a value of 20.
- So <code>result + 30</code> evaluates to 50 because <code>result</code> stores a value of 20 — <code>result + 30</code> becomes <code>20 + 30</code>.
- When we run <code>result = result + 30</code>, <code>result</code> is updated to store the output of <code>result + 30</code>, which is 50. Running <code>result = result + 30</code> is the same as running <code>result = 30 + result</code> because <code>result</code> stores 20.
- <code>print(result)</code> outputs 50 after we run <code>result = result + 30</code>.

6.String
===

***

6a. What is String
---

Thus far, we largely dealt with <code>int</code></b> and <code>float</code></b>, but in data analytics we use strings a lot as well. To create a <code>string</code> in python, simply store the data between a single/double quotation like the following:

````python
# Storing two string data, one in single quotation and another in double quotation
most_popular_game = "wii sports"
second_most_popular_game = 'Super Mario Bros'

# The output does not differ regardless whether you store them in single or double quotation
print(most_popular_game)
print(second_most_popular_game)

# They are both of string type regardless whether you store them in single or double quotation
print(type(most_popular_game))
print(type(second_most_popular_game))
````

6b. Double vs Single Quotation
---

However, we recommend that you use double-quotation for your strings, and reserve single quotation for when you need an apostrophe. Imagine if <b>Super Mario Bros</b> changed their name to <b>Super Mario's Bro</b>

````python
# You would get an error if you stuck to using single quotation for the whole string
second_most_popular_game = 'Super Mario's Bros'
````

The better approach is to use double-quotation for most string, and reserve your single quotation for the aspostrophes.

````python
second_most_popular_game = "Super Mario's Bros" #works perfectly fine
print(second_most_popular_game)
````

6c. Escape Character
---

To deal with the issue of quotation, you can also use an <b>escape character</b> instead of using double quotation for the entire string:

- \ (backslash is an escape character)

Basically by putting a backslash before special characters (those with special function) like quotation, it basically <b>treats the subsequent character as a literal character</b>, and ignores any special function it has.

````python
second_most_popular_game = 'Super Mario\'s Bros'
print(second_most_popular_game)
````

Try it again without the backslash and you'll realise that the code produces an error! This is because when python recognises the second single quotation as the end of the string, and hence it considered <b>Bros'</b> as not part of the string, thus giving an error. 

To work around this issue, you could either use the backslash character or <b>use double quotation instead</b> (illustrated below) if your string contains a single quotation character.

````python
second_most_popular_game = "Super Mario's Bros"
````
