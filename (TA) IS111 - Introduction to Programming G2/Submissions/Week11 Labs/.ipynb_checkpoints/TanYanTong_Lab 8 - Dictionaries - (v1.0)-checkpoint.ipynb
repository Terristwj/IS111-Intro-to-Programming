In this question, we consider a simple substitution cipher that replaces each letter in the original message with another letter.

### Part (a) [ &bigstar; ]

````python
#other solution
def encrypt(my_dict, msg):
    encrypted = ""
    for i in range(len(msg)):
        if msg[i] in my_dict:
            encrypted += my_dict[msg[i]]
            
        else:
            encrypted += msg[i]
            
    return encrypted
    
````
